# Model building, selection and validation

* For Model building,
  1. We leverage 3 different algorithms spanning across linear, tree-based and multi-layered dense networks: multivariate cox proportional-hazard with regularization (CPH-R), Random Survival Forests (RSF) [12], and neural networks.
  2. The neural networks model is built with core architecture, Neural Cox Non-Proportional-Hazard (NCNPH) which is based on relaxing the proportional hazard assumption (H Kvamme et al. 2019, JMLR [1]). 
  3. We leverage a Multistate (M) framework as described (by Gerstung et al. 2017, Nature Genetics [2]) taking inputs from the transition probabilities provided using NCNPH to get final probabilities adjusted for competing risks. Together, we call it <u> Individualized Risk Model for Myeloma (IRMMa)</u>. <br>
* For Model validation,
  1. We evaluate discriminative and calibrative performance.
  2. We evaluate the model's discrimation, by leveraging Harrell [3], Uno's[4] Concordance Index (C-index).
  3. We evaluate for calibration using Integrated Brier Score, Negative Binomial Log-Likelihood (NBLL). <br>
  4. We also show a clinically useful measure in the form of mean absolute difference (MAD). 
* To select the best model,
  1. We perform a 5-fold cross validation (CV) stratifying each split by Age, International Staging System (ISS), HDM-ASCT repeated 10 times in a random fashion.
  2. Wilcoxon rank-sum test [5] is used to calculate p-values.
  3. We evaluate risks at 5 years across [3, 4, 6, 7], where our risk model, IRRMa outperforms M-CPH-R, M-RSF. <br> 
* We introduce two variants of IRMMa:
  1. Vanilla, i.e., with all genomic features 
  2. With Top Genomics (TG), i.e., with genomic feature selection (see section 1.3.2). <br>
* We then compare existing risk stratification methods such as ISS, Revised ISS (R-ISS), R2-ISS with our model IRMMa (TG).
* Endpoints in this study are Overall Survival (OS), Event Free Survival (EFS) where the event is progression of disease (POD).
* IRMMa (TG) shows superior performance in comparison to the benchmark staging scoring systems in both prediction and calibration thus allowing us to build a more robust knowledge bank which we leverage in the rest of the literature. 

**Acronyms/definitions used in this literature:**<br>
* HDM-ASCT, ASCT and SCT are used interchangeably.
* Concordance and C-index are used interchangeably.
* Phase 1 (P1): induction treatment within 1 year, 
* Phase 2 (P2): addition of HDM-ASCT and/or continuous treatment or neither upon completion of P1 (without POD).
* Dec. : Deceased
* Internal: name of the cohort consisting of 5 datasets, MMRF CoMMpass (n=1062), Myeloma Genome Project (n=492), Moffitt Avatar study (n=177), UAMS (n=93) and MSKCC MyType (n=109).

````python
## Run python code to get files needed to run this section of code
def metric_rank_at_xyrs(sel_models, outcome_typ, __dir__, file_, metric_):
    '''
    sel_models (list of str) is to be ordered according to your expected output
    outcome_typ (str) is either os_cv, efs_cv, os_hd, efs_hd
    base_dir (str)
    file_ (str): all groups 
    metric_ (str): C_index_harrell, C_index_uno, brier score, nbll
    '''

    df_xyr = pd.read_csv(__dir__ + outcome_typ + "_" + str(time_risk_days) +
                         "_group_" + file_ + ".csv")[[metric_, 'X_ID', 'Fold']]
    df_ = pd.merge(df_xyr, rename_x_id_df, left_on='X_ID',
                   right_on='old')[['ID', metric_, 'Fold']]
    df_ = df_.dropna(subset=[metric_])
    df_ = df_[df_['ID'].isin(sel_models)]

    # rank models across folds
    df_sub_copy = df_.copy()
    df_sub = pd.pivot(df_, index='Fold', columns='ID', values=metric_)
    list_of_models = df_sub_copy.groupby(
        ['ID'])[metric_].median().to_frame().reset_index().set_index(
            'ID').reindex(sel_models).index.to_list()
    df_sub_c_ind_vals = df_sub.copy()

    if metric_ == 'Int_Brier_Score' or metric_ == 'NBLL':
        df_sub = df_sub
    elif metric_ == 'C_index_harrell' or metric_ == 'C_index_uno':
        df_sub = 1 - df_sub
    df_sub_rank = df_sub.T.rank(method='min')

    df_sub_rank = pd.DataFrame(
        df_sub_rank.T.value_counts().rename('count')).reset_index()

    concat_df = pd.DataFrame()  # percentage ranked order
    for item in list_of_models:
        df_model = df_sub_rank[[item, 'count']]
        df_model = df_model.groupby([item]).agg({
            'count': 'sum'
        }).rename(columns={'count': item})
        concat_df = pd.concat([concat_df, df_model], axis=1)

    concat_df = concat_df.fillna(0)
    concat_df = concat_df / concat_df.sum()[0]  #* 100
    concat_df.index = concat_df.index.astype(int)

    concat_df = concat_df.T
    return df_, concat_df, df_sub_c_ind_vals
````

````python
def label_diff(i, j, text, X, Y, ax):
    # Custom function to draw the diff bars
    x = (X[i] + X[j]) / 2
    y = 1.1 * max(Y[i], Y[j])
    dx = abs(X[i] - X[j])

    props = {
        'connectionstyle': 'bar',
        'arrowstyle': '-',
        'shrinkA': 20,
        'shrinkB': 20,
        'linewidth': 4,
        'family': 'Arial'
    }
    ax.annotate(text, xy=(x, y + 0.2), zorder=10)
    ax.annotate('', xy=(X[i], y), xytext=(X[j], y), arrowprops=props)


def barplot_annotate_brackets(num1,
                              num2,
                              data,
                              center,
                              he_ind,
                              height,
                              ax,
                              yerr=None,
                              dh=.03,
                              barh=.03,
                              fs=None,
                              maxasterix=None):
    if type(data) is str:
        text = data
    else:
        text = ''
        p = .05

        while data < p:
            text += '*'
            p /= 10.

            if maxasterix and len(text) == maxasterix:
                break

        if len(text) == 0:
            text = 'n. s.'

    lx, ly = center[num1], height[he_ind]
    rx, ry = center[num2], height[he_ind]

    if yerr:
        ly += yerr[num1]
        ry += yerr[num2]

    ax_y0, ax_y1 = plt.gca().get_ylim()
    dh *= (ax_y1 - ax_y0)
    barh *= (ax_y1 - ax_y0)

    y = max(ly, ry) + dh

    barx = [lx, lx, rx, rx]
    bary = [y, y + barh, y + barh, y]
    mid = ((lx + rx) / 2, y + barh)

    plt.plot(barx, bary, c='maroon')

    kwargs = dict(ha='center', va='bottom')
    if fs is not None:
        kwargs['fontsize'] = fs

    ax.text(*mid, text, **kwargs, fontsize=8)


def calc_pval(order_, a, b, df_metr_conc, item, metric, bars, he_ind, heights,
              ax):
    '''
    Calculate p-values using ranksums test
    '''
    ind_a = order_.index(a)
    ind_b = order_.index(b)

    stat, p_val = scipy.stats.ranksums(
        df_metr_conc[df_metr_conc[item] == a][metric],
        df_metr_conc[df_metr_conc[item] == b][metric])

    if p_val < 0.005:
        text = 'p<0.005'
    elif p_val < 0.001:
        text = 'p<0.001'
    elif p_val < 0.01:
        text = 'p<0.01'
    elif p_val < 0.05:
        text = 'p<0.05'
    elif p_val > 0.05:
        text = 'p=' + str(round(p_val, 2))

    barplot_annotate_brackets(ind_a, ind_b, text, bars, he_ind, heights, ax)
    return ax


def box_ranked_plot(title, metr_, df_, ranked_df_, order_, ax_box, ax_rank,
                    ytex, comparisons_, heights_):
    '''
    Box & stacked bar plot
    '''
    bar_width = 0.65
    bars = np.arange(len(order_))  # number of models
    he_ind = 0

    # box
    sns.boxplot(x='ID',
                y=metr_,
                palette='Set2',
                data=df_,
                notch=True,
                order=order_,
                width=bar_width,
                ax=ax_box,
                showfliers=True)
    ax_box.set_xlabel('')
    ax_box.set_ylabel(ytex)
    ax_box.set_title(title)

    if metr_ == 'C_index_harrell' or metr_ == 'C_index_uno':
        ax_box.set_ylim([0.45, 0.9])
    elif metr_ == 'Int_Brier_Score':
        ax_box.set_ylim([0, 0.35])
    elif metr_ == 'NBLL':
        ax_box.set_ylim([0.4, 1])

    mpl.pyplot.sca(ax_box)
    ax_box.axes.get_xaxis().set_visible(False)
    ax_box.spines['top'].set_visible(False)
    ax_box.spines['right'].set_visible(False)
    ax_box.spines['bottom'].set_visible(False)
    ax_box.spines['left'].set_color('black')
    ax_box.spines['left'].set_linewidth(1)

    '''
    for ind_, (a, b), in enumerate(comparisons_):
        calc_pval(order_, a, b, df_, 'ID', metr_, bars, he_ind, heights_,
                  ax_box)
        he_ind += 1
    '''
    # ranked
    ranked_df_[list(range(1, ranked_df_.shape[0] + 1))].plot(kind='bar',
                                                             width=bar_width,
                                                             ax=ax_rank,
                                                             stacked=True,
                                                             colormap='tab20')
    ax_rank.set_xticklabels(order_, rotation=60)
    ax_rank.set_ylim([0, 1])
    if title == 'OS':
        ax_rank.get_legend().remove()
        ax_rank.set_ylabel('Fraction of folds')
    else:
        ax_box.set_ylabel('')
        ax_rank.legend(loc='center left',
                       bbox_to_anchor=(1, 0.5),
                       frameon=False, title='Rank')

    mpl.pyplot.sca(ax_rank)
    ax_rank.spines['top'].set_visible(False)
    ax_rank.spines['right'].set_visible(False)
    ax_rank.spines['bottom'].set_visible(False)
    ax_rank.spines['left'].set_color('black')
    ax_rank.spines['left'].set_linewidth(1)

    return ax_box, ax_rank


def plot_metric_at_xyrs(order_, metr_, dfs_, ranked_dfs_, y_tex, comparisons_,
                        heights_):
    '''
    Called
    '''
    fig, axs = plt.subplots(2, 2)

    ax0, ax1, ax2, ax3 = axs.flatten()

    box_ranked_plot('OS', metr_, dfs_[0], ranked_dfs_[0], order_, ax0, ax2,
                    y_tex, comparisons_, heights_[0])
    box_ranked_plot('EFS', metr_, dfs_[1], ranked_dfs_[1], order_, ax1, ax3,
                    y_tex, comparisons_, heights_[1])
    plt.tight_layout()
    return fig
````

## Discriminative performance

* The Discriminative performance calculates concordance ranked across sets of patients comparing real and observed probabilities. 
* We leverage two metrics as provided by Harrell and Uno et al.:
  1. Harrell's Concordance (C) is a goodness of fit measure of models that produce risk scores, and is the most  used on all pairs of patients.
  2. Uno's Concordance (C) considers right censoring of pairs of patients in calculating concordance, based on inverse weighting of probability of censoring distributions.
* In this section, we show three figures for each metric:
  1. Comparing existing benchmark algorithms M-CPH-R, M-RSF, IRMMa (Vanilla), IRMMa (TG) shows the best performance. 
  2. Comparing existing benchmark risk models in Myeloma: ISS, R-ISS, R2-ISS with IRMMa (TG).
  3. Concordance at 1, 2, 3, 4, 5 years from start of induction.
* Within each figure, we show concordance and ranking of the model at each cross validated split.
* In summary, IRMMa (TG) performs the best in concordance both with and without considering for censoring.

## Calibrative performance

* To measure the accuracy of survival models, we can leverage commonly used methods such as Integrated Brier Score (BS) and Negative Binomial Log-likelihood (NBLL). 
  1. BS is a binary classification metric that measures calibration based on inverse probability of censoring weighted (IPCW). The lower the score the better with 0 indicating perfect accuracy.
  2. NBLL measures both discrimination and calibration and is the negative of the binomial log-likelihood based on the IPCW. Lower value indicates better balance.
* In this section, we show three figures for each metric:
  1. Comparing existing benchmark risk models in Myeloma: ISS, R-ISS, R2-ISS with IRMMa (TG).
  2. Evaluating at 1, 2, 3, 4, 5 years from start of induction.
* Within each figure, we show concordance and ranking of the models at each cross validated split.
* Our model IRMMa (TG) has the best calibration in terms of BS at less than 14% and 18% for OS and EFS at 5 years respectively. 
* Balancing for discriminative and calibrative measures, i.e., in terms of NBLL, IRMMa (TG) shows the best performance of ~50% and <60% for OS and EFS respectively.

## Feature ranking and selection

* Many methods of feature ranking exist. From greedy stepwise linear forward/backward approaches to model-agnostic approaches such as permutation importance [8] which inspect the features of the model, to frameworks for interpreting model predictions such as LIME [9], and SHapley Additive exPlanations (SHAP) [10]. 
* We utilize IRMMa to rank inidividual features in P1/P2 using permutation importance technique (shuffled n=5 for a feature).
* We show below in figure, a ranked barplot where the mean and variance are determined by aggregating the values in each shuffle (5 times) and cross-validation split (total 50 splits), i.e. 5x50 = 250 values.
* It is also important to note that the permutation importance technique is not able to indicate the true intrinsic value of a feature, but how well it is for a model.
* In this section we will present: 
    1. Feature ranking using permutation importance of the different OS/EFS states in P1/P2.
    2. Genomic feature selection
* For genomic feature selection, rather than fully conservative or elaborate search through a relatively large feature space, we explore a hybrid approach (see 1.3.2).

### Permutation importance based feature ranking

````python
model = 'neural_cox_non_prop'
output_files = glob.glob(
    base_dir + "/data_out/expt_1_matrix_04052023/permut_feat_imps/compose/*_all_feat_*" +
    model + "*.csv")
data_list = []
for filename in output_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    data_list.append(df)
    df['fold'] = filename.split('/')[-1].split('_')[1]

concat_feats = pd.concat(data_list, axis=0, ignore_index=True)

concat_feats = concat_feats[[
    'group', 'class_generic', 'weight', 'std', 'fold', 'column_id'
]]

overall_fea_class = pd.read_excel(base_dir + '/data_in/legend_PMMM_2023.xlsx')
concat_feats = pd.merge(concat_feats,
                        overall_fea_class[['column_id', 'display_id']],
                        on='column_id',
                        how='inner')

unique_feat_groups = concat_feats['class_generic'].unique().tolist()
color = ['deepskyblue', 'pink', 'orange', 'forestgreen', 'violet', 'gray']
concat_feats['color'] = ''
for i, item in enumerate(unique_feat_groups):
    # print(item, color[i])
    concat_feats['color'].mask(concat_feats['class_generic'] == item,
                               color[i],
                               inplace=True)

concat_feats['display_id'] = concat_feats['display_id'].str.replace(
    r'(CNV_)|(CNV\.SNV_)|(SNV_)', '', regex=True)

unique_states = [
    'Progress (P1)', 'Progress & deceased (P1)', 'Progress (P2)',
    'Progress & deceased (P2)'
] 
````

### Genomic Feature selection

***
* Here we define IRMMa (TG) and provide pseudo-code to get the final set of genomic features at 5 years.
* By integrating univariate features which are further refined and sub-selected by our model, we shortlist 20 genomic features. We define this as our hybrid approach. <br> 
***
**Below is the pseudo-code:** 
***
```
define univ_feats_p_less_than_01 = choose univariate genomic features with p-val <0.01 for OS and EFS 
define univ_feats_p_less_than_05 = choose univariate genomic features with p-val <0.05 for OS and EFS 
define baseline_features = ISS+ Clinical+ Recurrent translocation+ HDM-ASCT+ Continuous treatment+ Therapy
pool_to_select = univ_feats_p_less_than_05 - univ_feats_p_less_than_01
rank pool_to_select based on permutation ranking in P1/P2

cumul_feature = empty list
```

````python
def clean_transform_raw_data(df):
    df_cp = df.copy()

    df_cp.drop(['time_SCT', 'duration', 'study', 'DARA', 'ELO', 'SCT_line'], axis=1, inplace=True)

    df_feats = df_cp
    # map clinical vars
    map_gender = {'male': 0, 'female': 1, 'Male': 0, 'Female': 1}
    df_feats = df_feats.replace({'gender': map_gender})

    map_ecog = {'ecog<2': 0, 'ecog>=2': 1}
    df_feats = df_feats.replace({'ecog': map_ecog})

    map_iss = {'ISS1': 0, 'ISS2': 1, 'ISS3': 2, 'I': 0, 'II': 1, 'III': 2}
    df_feats = df_feats.replace({'ISS': map_iss})

    map_r_iss = {'R-ISS1': 0, 'R-ISS2': 1, 'R-ISS3': 2}
    df_feats = df_feats.replace({'R_ISS': map_r_iss})

    map_r2_iss = {'I': 0, 'II': 1, 'III': 2, 'IV':3}
    df_feats = df_feats.replace({'R2_ISS': map_r2_iss})

    df_feats["combo"] = np.select([df_feats.combo <= 1, df_feats.combo > 1], [0, 1])

    map_ldh = {'Low': 0, 'High': 1, 'Normal': 2}
    df_feats = df_feats.replace({'LDH_level': map_ldh})

    map_race = {'WHITE': 0, 'BLACK OR AFRICAN AMERICAN': 1, 'OTHER': 2}
    df_feats = df_feats.replace({'Race': map_race})

    df_feats["age_group"] = np.select([(df_feats.age < 65), (df_feats.age >= 65) & (df_feats.age < 75), df_feats.age >= 75], [0, 1, 2])

    df_feats['phases'] = np.where((df_feats['phase'] == 'Phase1-only'), 'phase_1', 'phase_2')
    df_feats.drop('phase', axis=1, inplace=True)
    
    df_feats = df_feats.set_index('sample')

    return df_feats
````

````python
def data_normalize(x_array, normaliz_type):
    """
    :param normaliz_type: data normalize / scalar technique
    :return: preprocessed numpy arrays x,y to feed into ML
    """

    if normaliz_type == 'normalizer':
        scaler = preprocess.Normalizer().fit(x_array)

    # Approach 2 Min-Max
    if normaliz_type == 'min_max':
        # print('Normalization using min-max')
        scaler = preprocess.MinMaxScaler(feature_range=(0, 1))

    if normaliz_type == 'max_abs':
        # print('Max abs scale')
        scaler = preprocess.MaxAbsScaler()

    if normaliz_type == 'std_scaler':
        # print('Standard scaler')
        scaler = preprocess.StandardScaler()

    if normaliz_type == 'power_transformer':
        # print('Power transformer')
        scaler = preprocess.power_transform(method='yeo-johnson')

    if normaliz_type is None:
        # print('No normalization performed')
        return x_array, None

    if x_array.shape[1] != 1:
        x_norm = scaler.fit_transform(x_array)
    else:
        x_norm = scaler.fit_transform(x_array.reshape(-1, 1))
    return x_norm, scaler

def imputer_knn(df):
    x_array = df.values
    x_array, _ = data_normalize(x_array, 'min_max')
    imputer = KNNImputer(n_neighbors=5, weights='uniform', metric='nan_euclidean')
    imputer.fit(x_array)
    x_imputed = imputer.transform(x_array)
    # print('Missing: %d' % sum(np.isnan(x_imputed).flatten()))
    return x_imputed, imputer
````

# Knowledge bank

* To construct a bank of information built on the 1933 patients (internal cohort) used, we leveraged the risks generated from IRMMa (TG). For every new test case, we first impute missing values using k-nearest neighbor-based imputation followed by a fast k-nearest neighbor search called k-d tree [11] and aggregate the median across the top-k nearest patients that have received the same treatment and other clinical or genomic attributes. 
* We provide an interactive tool via a website application for research purposes of the community: https://irmma-risk-calculator.miami.edu. For any given user query, we build real-time k-d tree, a fast unsupervised nearest neighbor’s approach to find nearest neighboring patients from our knowledge bank that are identical in ISS, treatments and are in the age range (<65, 65-74, >=75) and have similar clinical and genomics features after having passed through weighted patient filtering based on Jaccard distance. We then report the median risk(s) across the top-k (<= 10 patients) for the multistate and treatment variance plots. 
* If user query has consolidation / maintenance, the model picks from patients who have progressed in P1 having same user provided therapy and in P2, having same consolidation / maintenance and user given therapy. If no consolidation / maintenance, choose from patients in P1/P2.

**In this section:**<br>
1. Firstly, we present the knowledge bank pseudo-code of the overall aproach and the core algorithm used for retrieving patients with similar characteristics to a query (on the website).
2. Secondly, we see overlap of risks of patients with HDM-ASCT and (less than 70 years old) between internal  and GMMG-HD6 trial cohorts.
3. Lastly, we show the OS and EFS risk distributions annotated with the ground truth event.

**The pseudo-code for the knowledge bank is as follows,** <br> 
```
Prepare test sample:
    Train K-Nearest Neighbors (KNN) imputer(on internal cohort) including the test sample.
    Apply trained KNN imputer on new sample (from GMMG-HD6 or user query)

Build a real-time k-dimensional (k-d) tree on knowledge bank
For every new test sample,
    input to k-d tree to get nearest (<=10) patients to & report median risk(s)
```

````python
def nearest_samples_by_treatment(query_int_dict, query_clin_gen_vals, feat_ids, treats, df_raw_internal_cp):
    ## weighted knn

    # treats = ['chemo', 'KAR', 'LEN', 'BORT', 'THAL', 'SCT_first_line', 'continuos_treat']
    # select patients with same treatment combinations as query
    if (query_int_dict[treats[5]] == 1 | query_int_dict[treats[6]] == 1):
        # print('\n\nif consolidation then also choose patients that prog./dec in P1 in addition to matched P2')
        df_p1_match = df_raw_internal_cp[
            (df_raw_internal_cp[treats[0]] == query_int_dict[treats[0]]) &
            (df_raw_internal_cp[treats[1]] == query_int_dict[treats[1]]) & 
            (df_raw_internal_cp[treats[2]] == query_int_dict[treats[2]]) &
            (df_raw_internal_cp[treats[3]] == query_int_dict[treats[3]]) &
            (df_raw_internal_cp[treats[4]] == query_int_dict[treats[4]]) &
            (df_raw_internal_cp['phases'] == 'phase_1') &
            (df_raw_internal_cp['pfs_code'] == 1)] # & (df_raw_internal_cp['os_code'] == 1)

        df_p2_match = df_raw_internal_cp[
            (df_raw_internal_cp[treats[0]] == query_int_dict[treats[0]]) &
            (df_raw_internal_cp[treats[1]] == query_int_dict[treats[1]]) & 
            (df_raw_internal_cp[treats[2]] == query_int_dict[treats[2]]) &
            (df_raw_internal_cp[treats[3]] == query_int_dict[treats[3]]) &
            (df_raw_internal_cp[treats[4]] == query_int_dict[treats[4]]) &
            (df_raw_internal_cp[treats[5]] == query_int_dict[treats[5]]) &
            (df_raw_internal_cp[treats[6]] == query_int_dict[treats[6]]) &
            (df_raw_internal_cp['phases'] == 'phase_2')]
        df_raw_internal_cp = df_p1_match.append(df_p2_match, ignore_index=False)
        # print(df_p1_match.shape, df_p2_match.shape, df_raw_internal_cp.shape)
    else:
        # print('\n\nif no consolidation select from all patients including P1')
        df_p1_or_p2 = df_raw_internal_cp[
            (df_raw_internal_cp[treats[0]] == query_int_dict[treats[0]]) &
            (df_raw_internal_cp[treats[1]] == query_int_dict[treats[1]]) & 
            (df_raw_internal_cp[treats[2]] == query_int_dict[treats[2]]) &
            (df_raw_internal_cp[treats[3]] == query_int_dict[treats[3]]) &
            (df_raw_internal_cp[treats[4]] == query_int_dict[treats[4]]) &
            (df_raw_internal_cp[treats[5]] == query_int_dict[treats[5]]) &
            (df_raw_internal_cp[treats[6]] == query_int_dict[treats[6]])]
        df_raw_internal_cp = df_p1_or_p2
        # print(df_p1_or_p2.shape, df_raw_internal_cp.shape)

    ####### find similar patients according to clinical, genomics/translocations
    df_raw_internal_cp = df_raw_internal_cp.dropna(subset=feat_ids)
    # print('\nbefore tree search shape of df after removing rows of knowledge bank with na', df_raw_internal_cp.shape)
    
    # filter by age & ISS of knoweledge bank dataset candidates wrt. query
    df_raw_internal_cp = df_raw_internal_cp[(df_raw_internal_cp['age']==query_int_dict['age']) &
                                (df_raw_internal_cp['ISS']==query_int_dict['ISS'])]

    # print('\nfilter by age and ISS', df_raw_internal_cp.shape)

    # impute knowledge bank dataset - IRMMa (TG) except age, ISS, treatment
    try:
        train_x_clin_gen, _ = imputer_knn(df_raw_internal_cp[set(feat_ids)-set(treats)-set(['age','ISS','phases', 'pfs_code', 'os_code'])])
        # print('data available cos - shape is', df_raw_internal_cp.shape)
    except Exception as e:
        # print('no data returned cos - shape is', df_raw_internal_cp.shape)
        return []

    ## normalize training set
    train_x_clin_gen, norm = data_normalize(train_x_clin_gen, 'min_max')
    query_clin_gen_vals = norm.transform(query_clin_gen_vals.to_numpy().reshape(1,-1))
    from scipy.spatial import distance

    del_ind = []
    for ind, row in enumerate(train_x_clin_gen):
        d = distance.jaccard(row, query_clin_gen_vals)
        #print('\nq1', query_clin_gen_vals) 
        #print('tS with row', ind) 
        #print(row)
        # print('distance', d)
        # print('number of mismatches', np.sum(query_clin_gen_vals != row))
        if (d > 0.7) & (np.sum(query_clin_gen_vals != row) > 12):
            del_ind.append(ind)
    
    # print('before removing higher distance samples', train_x_clin_gen.shape)
    
    train_x_clin_gen = np.delete(train_x_clin_gen, del_ind, axis=0)

    # print('after removing higher distance samples', train_x_clin_gen.shape, '\n')

    ###### build tree if atleast 3 training samples available
    if train_x_clin_gen.shape[0] > 1:
        tree_obj_clin_gen = KDTree(train_x_clin_gen)
    else:
        return []

    ###### search for similar patients wrt. query
    if not np.isnan(query_clin_gen_vals).any():

        # varying top_k based on number of candidates to choose from 
        if train_x_clin_gen.shape[0] > 10:
            top_k = 10
        else:
            top_k = train_x_clin_gen.shape[0]

        dist, indices = tree_obj_clin_gen.query(query_clin_gen_vals, k=top_k)
        #print('\n\nafter tree search closest distances and indices - based on iss clin gen\n', dist, indices)

        retriev_sample_ids = []
        for (dist_, ind_) in zip(dist[0], indices[0]):
            if dist_ == 0: # conditional to closest neigbors
                #print('\nfound zero distance neigbors')
                sample_id = df_raw_internal_cp[set(feat_ids)-set(treats)].iloc[ind_].name
                retriev_sample_ids.append(sample_id)

        if len(retriev_sample_ids) == 0:
            #print('\nzero distance neigbors not found, default to other nearest ones')
            for (dist, ind) in zip(dist[0], indices[0]):
                sample_id = df_raw_internal_cp[set(feat_ids)-set(treats)].iloc[ind].name
                retriev_sample_ids.append(sample_id)
        else:
            print('Zero distance neigbors found in our dataset')

        #print('\n\nclosest sample ids\n\n', retriev_sample_ids)
        
        return retriev_sample_ids
````

````python
def knowledge_bank(query):
    # start = time.time()

    # load/read
    df_raw_internal = pd.read_csv(targ_dir + '/website/df_raw_internal.csv', index_col='sample')
    df_raw_hd6 = pd.read_csv(targ_dir + '/website/df_raw_hd6.csv', index_col='sample')
    to_be_mapped_header = pd.read_csv(targ_dir + '/website/to_be_mapped_header.csv')['0'].tolist()
    df_treat_adj_risk = pd.read_csv(targ_dir + 'website/df_treat_adj_risk.csv')
    loo_orig_files = pd.read_csv(targ_dir + 'website/loo_orig_filenames.csv')
    imputer = pickle.load(open(targ_dir + 'website/imp_knn.pkl', 'rb')) 
    
    df_raw_internal_cp = df_raw_internal.copy()
    df_raw_internal_cp['age'] = np.select([(df_raw_internal_cp.age < 65), (df_raw_internal_cp.age >= 65) & (df_raw_internal_cp.age < 75), df_raw_internal_cp.age >= 75], [0, 1, 2])
    
    df_raw_hd6_cp = df_raw_hd6.copy()
    df_raw_hd6_cp['age'] = np.select([(df_raw_hd6_cp.age < 65), (df_raw_hd6_cp.age >= 65) & (df_raw_hd6_cp.age < 75), df_raw_hd6_cp.age >= 75], [0, 1, 2])
    
    # phase 1 & 2 treats
    treats = ['chemo', 'KAR', 'LEN', 'BORT', 'THAL', 'SCT_first_line', 'continuos_treat']
    
    query_feat_vals = df_raw_hd6_cp.loc[query][to_be_mapped_header]
    query_int_dict = query_feat_vals.to_dict()

    # print(query, query_int_dict, '\n')

    if query_int_dict['age'] >= 75:
        query_int_dict['age'] = 2
    elif (query_int_dict['age'] >= 65) & (query_int_dict['age'] < 75):
        query_int_dict['age'] = 1
    else:
        query_int_dict['age'] = 0
    
    feat_ids = list(query_int_dict.keys())

    # impute query
    test_x_ = imputer.transform(query_feat_vals.to_numpy().reshape(1, -1))
    query_feat_vals = pd.DataFrame(data=test_x_)
    query_feat_vals.columns = to_be_mapped_header
    query_feat_vals = query_feat_vals.squeeze(axis=0)

    query_clin_gen_vals = query_feat_vals[set(feat_ids)-set(treats)-set(['age'])-set(['ISS'])-set(['phases'])]
    query_clin_gen_vals = query_clin_gen_vals.round()
    # print('\n\nafter rounding imputed query- clin, gen values\n', query_clin_gen_vals)

    # treat combos available in our dataset 
    avail_treats_internal = ['chemo,KAR', 'chemo,KAR,LEN', 'chemo,BORT', 'chemo,LEN', 'LEN', 'chemo',
                             'chemo,THAL', 'chemo,LEN,BORT', 'KAR,LEN', 'LEN,BORT']

    avail_p2_internal = ['', 'HDM-ASCT', 'c-treatment', ['HDM-ASCT', 'c-treatment']]
    
    # user chosen P1/P2 treatment
    user_chosen_treats_x = []
    for item in treats[:5]:
        if query_int_dict[item] == 1:
            user_chosen_treats_x.append(item)
            
    ## consol / maintainence
    user_consol_temp = []
    for item in treats[5:]:
        if query_int_dict[item] == 1:
            user_consol_temp.append(item)

    count = 0

    user_consol = []
    for item in user_consol_temp:
        if item == 'SCT_first_line':
            user_consol.append('HDM-ASCT') 
        elif item == 'continuos_treat':
            user_consol.append('c-treatment')

    # print('\n\nuser chosen - age ISS SCT Cont combo', query_int_dict['age'], query_int_dict['ISS'], query_int_dict['SCT_first_line'], query_int_dict['continuos_treat'], query_int_dict['combo'], user_chosen_treats_x, user_consol, '\n')

    # induction/consol combo available
    for p1_treats in avail_treats_internal:
        if collections.Counter(user_chosen_treats_x) == collections.Counter(p1_treats.split(',')):
            avail_p1 = True
    for p2_treats in avail_p2_internal:
        if (p2_treats == '') or (p2_treats == 'HDM-ASCT') or (p2_treats == 'c-treatment') or (collections.Counter(p2_treats) == collections.Counter(['c-treatment', 'HDM-ASCT'])):
            avail_p2 = True
    try:
        avail_p1 == True
        avail_p2 == True
    except Exception as e:
        # print('exception again')
        return query, pd.DataFrame(), pd.DataFrame()
    
    # save treatment var
    df_treat_var = pd.DataFrame()
    
    query_int_dict_cp = query_int_dict.copy()
    for treat_combo, treat_consol in list(it.product(avail_treats_internal, [['c-treatment'], ['HDM-ASCT'], [''], ['HDM-ASCT', 'c-treatment']])):
        # P1
        indiv_treats = treat_combo.split(',')
        for item in indiv_treats:
            query_int_dict_cp[item] = 1
        for item in list(set(treats[:5])-set(indiv_treats)):
            query_int_dict_cp[item] = 0

        # consolidation
        if treat_consol == ['']:
            sct_cont_state = 'P2_no_SCT_no_cont'
            query_int_dict_cp['SCT_first_line'] = 0
            query_int_dict_cp['continuos_treat'] = 0
        elif treat_consol == ['HDM-ASCT']:
            sct_cont_state = 'P2_SCT_only'
            query_int_dict_cp['SCT_first_line'] = 1
            query_int_dict_cp['continuos_treat'] = 0
        elif treat_consol == ['c-treatment']:
            sct_cont_state = 'P2_cont_only'
            query_int_dict_cp['SCT_first_line'] = 0
            query_int_dict_cp['continuos_treat'] = 1
        elif collections.Counter(treat_consol) == collections.Counter(['c-treatment', 'HDM-ASCT']):
            sct_cont_state = 'P2_SCT_cont'
            query_int_dict_cp['SCT_first_line'] = 1
            query_int_dict_cp['continuos_treat'] = 1

        # print(indiv_treats, sct_cont_state)
        sample_ids_by_treat = nearest_samples_by_treatment(query_int_dict_cp, query_clin_gen_vals, feat_ids, treats, df_raw_internal_cp)
        # print('sample ids len', len(sample_ids_by_treat), sample_ids_by_treat, '\n')

        # search enabled if we have more than x patient to aggregate on
        if len(sample_ids_by_treat) > 1:

            ####### if user chosen treatment - plotA
            if (collections.Counter(user_chosen_treats_x) == collections.Counter(indiv_treats)) and (collections.Counter(user_consol) == collections.Counter(treat_consol)):
                # print('yes - get plot1')
                ##### (1) at multiple time points - read files of retrieved sample ids (from /loo/) #####
                data_list = []
                for k_file_id in sample_ids_by_treat:
                    filename = loo_orig_files[loo_orig_files['0'].str.contains(k_file_id)]
                    single_df = pd.read_csv(filename.values[0][0], index_col=None, header=0)
                    data_list.append(single_df.head(1825))

                df = pd.concat(data_list, axis=1)
                
                df = df.drop(['sample'], axis=1)
                # df = df.drop(['sample', 'Time (Days)'], axis=1)
                
                # print('df', df.shape)
                
                #### median
                df_accum = pd.DataFrame()
                for ind, group in enumerate(df.columns.unique()):
                    spec_cols = [col for col in df.columns if group == col]
                    df_spec = df[spec_cols]
                    df_median = df_spec.median(axis=1).abs()
                    df_median = df_median.reset_index()
                    df_median = df_median.drop('index', axis=1)
                    df_median.columns = [group]
                    df_accum = pd.concat((df_accum, df_median), axis=1)

                # print(df_accum)
                df_accum = df_accum.rename(columns={'Alive in Induction': 'Alive in P1', 
                    'Alive in POD (induction)': 'Alive after POD (P1)', 
                    'Alive in Phase 2': 'Alive in P2',
                        'Alive in POD': 'Alive after POD (P2)', 
                            'Death in POD (induction)': 'Death after POD (P1)',
                                'Death in Phase 2': 'Death in P2', 
                                    'Death in POD (Phase 2)': 'Death after POD (P2)'})

                # print('first\n', df_accum.shape)
                # print('\n', df_accum.iloc[1824])

            ####### (2) plot B - treatment adj risk at 5 years #######
            sub_retr_ids = df_treat_adj_risk[df_treat_adj_risk['sample_id'].isin(sample_ids_by_treat)][['state', 'treatment', 'risk_pfs']]
            # select rows of patients from our bank who got same first-line and consolidation strategy as query
            sub_retr_ids = sub_retr_ids[(sub_retr_ids['state'] == sct_cont_state) & (sub_retr_ids['treatment'] ==treat_combo)]
            sub_retr_ids = sub_retr_ids.groupby(['state', 'treatment']).agg(risk_pfs=('risk_pfs', 'median'))
            sub_retr_ids = sub_retr_ids.round(3)
            
            df_treat_var = pd.concat([df_treat_var, sub_retr_ids])
            count +=1
            
        else:
            # print('less than x patients - so do not aggregate - value is NA')
            df_treat_var = pd.concat([df_treat_var, pd.DataFrame({'risk_pfs': np.nan}, index=[(sct_cont_state, treat_combo)])])
    
    try:
        len(df_accum) == 0
    except Exception as e:
        #print('exception')
        #print('age ISS SCT Cont', query_int_dict['age'], query_int_dict['ISS'], query_int_dict['SCT_first_line'], query_int_dict['continuos_treat'], user_chosen_treats_x, user_consol, '\n')
        return query, pd.DataFrame(), pd.DataFrame()
    
    ## plotB
    df_treat_var.index = pd.MultiIndex.from_tuples(df_treat_var.index, names=["state", "treatment"])
    df_treat_var.reset_index(inplace=True)
    
    # median pfs risk
    df_treat_var = pd.pivot_table(df_treat_var, values="risk_pfs", index=["treatment"], columns=["state"])

    df_treat_var = df_treat_var.rename(columns={'P2_no_SCT_no_cont': 'No ASCT & No C.T.',
                                              'P2_SCT_only' : 'ASCT', 'P2_cont_only': 'C.T.',
                                              'P2_SCT_cont': 'ASCT & C.T.'})
    
    df_treat_medians_var = df_treat_var.copy()
    # df_treat_var = df_treat_var[['No ASCT & No C.T.', 'C.T.', 'ASCT', 'ASCT & C.T.']]
    # df_treat_var = [list(df_treat_var)] + df_treat_var.reset_index().values.tolist()

    if ('HDM-ASCT' not in user_consol) & ('c-treatment' not in user_consol):
        y = 'No ASCT & No C.T.'
    elif ('HDM-ASCT' not in user_consol) & ('c-treatment' in user_consol):
        y = 'C.T.'
    elif ('HDM-ASCT' in user_consol) & ('c-treatment' not in user_consol):
        y = 'ASCT'
    elif ('HDM-ASCT' in user_consol) & ('c-treatment' in user_consol):
        y = 'ASCT & C.T.'

    # print(df_treat_medians_var.columns)
    df_treat_medians_var['list_treats'] = df_treat_medians_var.index.str.split(',')
    ref_pfs_risk = df_treat_medians_var[df_treat_medians_var.apply(lambda row: set(row['list_treats']) == set(user_chosen_treats_x), axis=1)].loc[:,y]
    df_treat_medians_var.drop('list_treats', axis=1, inplace=True)

    # df_treat_medians_var = df_treat_medians_var[['No ASCT & No C.T.', 'C.T.', 'ASCT', 'ASCT & C.T.']]
    df_treat_medians_var = df_treat_medians_var.stack(dropna=False).reset_index().rename(columns={0: 'risk_pfs'})
    df_treat_medians_var['sample'] = query
    
    # df_accum = df_accum.iloc[1824]
    # df_accum = pd.DataFrame(df_accum).T
    #df_accum = df_accum.rename(index={1824:query})
    df_accum['sample'] = query
    # print(df_accum)

    # reset
    query_int_dict_cp = query_int_dict.copy()

    # end = time.time()
    # print('\nsample time', end - start)
    # print(count)

    return query, df_accum, df_treat_medians_var
````
